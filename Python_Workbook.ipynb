```python
#removing space from column name
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
#creating the ranks for income
ranks = [0, 100000, 300000, float('inf')]
bands = ['Low', 'Medium', 'High']

df['Income_ranking'] = pd.cut(df['Estimated_Income'], bins=ranks, labels=bands, right=False)
```

```python
print(df['Income_ranking'].value_counts())
df['Income_ranking'].value_counts().plot(kind='bar')
```

```python
#doing some categorical columns analysis(sample 3 chosen)

categorical_columns = df [["BRId", "GenderId", "IAId"]].columns

for i in categorical_columns:
    print(df[i].value_counts(),end='\n')
    df[i].value_counts().plot(kind='bar')
```

```python
#univarite analysis

for i, predictor in enumerate(df[["BRId", "GenderId", "IAId"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
    
```

```python
#bivarite analysis

for i, predictor in enumerate(df[["BRId", "GenderId", "IAId"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='GenderId')
```

```python
#histogram plot
for col in categorical_columns:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title("Histogram Plot")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
#numerical column analysis(sample 3 chosen)

numerical_columns = ['Estimated_Income', 'Superannuation_Savings', 'Business_Lending']

for i, col in enumerate(numerical_columns):
    plt.figure(figsize=(20,17))
    plt.subplot(3,1,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
```

```python
#doing heatmaps for other sample sets of numerical data columns

numerical_columns2 = ['Credit_Card_Balance', 'Bank_Loans', 'Bank_Deposits', 'Saving_Accounts']
corelation_matrix = df[numerical_columns2].corr()

plt.figure(figsize=(10,13))
sns.heatmap(corelation_matrix, annot=True, cmap='crest', fmt='.3f')
plt.title('Heatmap')
plt.show()
```
